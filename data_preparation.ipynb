# Dataset preparation
Dataset source:
[link](https://www.eea.europa.eu/data-and-maps/data/aqereporting-9)

Dataset will combine data of PM10, PM2.5, CO, C6H6, SO2 and NO2 measurements in years 2018-2020 (measured in Bratislava, Slovakia).

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

Start by combining data of respective years and later appending them together into one csv file. First is data from 2018.

```python
df_2018 = pd.read_csv('raw_data/2018/SK_5_27208_2018_timeseries_PM10.csv')
df_2018.head()
```

```python
df_2018.rename(columns={'Concentration': 'PM10 Concentration', 'UnitOfMeasurement': 'PM10 Unit'}, inplace=True)
```

```python
df_pm2_5_2018 = pd.read_csv('raw_data/2018/SK_6001_35900_2018_timeseries_PM25.csv')
df_pm2_5_2018.rename(columns={'Concentration': 'PM2.5 Concentration', 'UnitOfMeasurement': 'PM2.5 Unit'}, inplace=True)
df_pm2_5_2018 = df_pm2_5_2018[['DatetimeBegin', 'DatetimeEnd', 'PM2.5 Concentration', 'PM2.5 Unit']]
df_2018 = pd.merge(df_2018, df_pm2_5_2018, how='outer')
```

```python
df_co_2018 = pd.read_csv('raw_data/2018/SK_10_27210_2018_timeseries_CO.csv')
df_co_2018.rename(columns={'Concentration': 'CO Concentration', 'UnitOfMeasurement': 'CO Unit'}, inplace=True)
df_co_2018 = df_co_2018[['DatetimeBegin', 'DatetimeEnd', 'CO Concentration', 'CO Unit']]
df_2018 = pd.merge(df_2018, df_co_2018, how='outer')
```

```python
df_no2_2018 = pd.read_csv('raw_data/2018/SK_8_27323_2018_timeseries_NO2.csv')
df_no2_2018.rename(columns={'Concentration': 'NO2 Concentration', 'UnitOfMeasurement': 'NO2 Unit'}, inplace=True)
df_no2_2018 = df_no2_2018[['DatetimeBegin', 'DatetimeEnd', 'NO2 Concentration', 'NO2 Unit']]
df_2018 = pd.merge(df_2018, df_no2_2018, how='outer')
```

```python
df_so2_2018 = pd.read_csv('raw_data/2018/SK_1_27236_2018_timeseries_SO2.csv')
df_so2_2018.rename(columns={'Concentration': 'SO2 Concentration', 'UnitOfMeasurement': 'SO2 Unit'}, inplace=True)
df_so2_2018 = df_so2_2018[['DatetimeBegin', 'DatetimeEnd', 'SO2 Concentration', 'SO2 Unit']]
df_2018 = pd.merge(df_2018, df_so2_2018, how='outer')
```

```python
df_c6h6_2018 = pd.read_csv('raw_data/2018/SK_20_27277_2018_timeseries_C6H6.csv')
df_c6h6_2018.rename(columns={'Concentration': 'C6H6 Concentration', 'UnitOfMeasurement': 'C6H6 Unit'}, inplace=True)
df_c6h6_2018 = df_c6h6_2018[['DatetimeBegin', 'DatetimeEnd', 'C6H6 Concentration', 'C6H6 Unit']]
df_2018 = pd.merge(df_2018, df_c6h6_2018, how='outer')
```

```python
df_2018.drop(columns=['Countrycode', 'Namespace', 'AirQualityNetwork', 'AirQualityStation', 'SamplingPoint', 'SamplingProcess', 'Sample', 'AveragingTime', 'AirPollutantCode', 'AirPollutant', 'AirQualityStationEoICode', 'Validity', 'Verification'], inplace=True)
df_2018
```
